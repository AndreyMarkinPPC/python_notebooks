## Change column from one type to another with `.astype("type")` method

For instance `df["col"] = df["col"].astype("int64")`

```python
# let's change the target metric `churn` from bool to int64
df.Churn = df["Churn"].astype("int64")
df.head()
```

Get statistical information with `.describe()` method

```python
df.describe()
```

By default `.describe()` shows information only on numerical features, 
to include info on non-numeric as well pass type of elements 
you would like to get stats for - `describe(include = ["object", "bool"])

```python
df.describe(include = ["bool", "object"])
```

For categorical features (type `object`) and boolean (type `bool`) we can use method
`value_counts()` to get counts.

```python
df["Churn"].value_counts()
```

By passing `normalize = True` to `value_counts()` we can get not counts but proportions

```python
df.Churn.value_counts(normalize = True)
```

# Applying functions to cells, columns and rows with `.apply()`

```python
# find max of each numeric column
df.apply(np.max)
```

By default `.apply()` works on rows, to apply to column pass `(axis = 1)` as an argument

```python
# Works with lambdas

# find information on all states which starts with 'W'
df[df.State.apply(lambda state: state[0] == "W")].head(15)
```

## Replace values in a column with `.map()` which requires a dictionary for replacement

```python
# create dictionary
d = {"No": False, 
    "Yes" : True}

df["International plan"] = df["International plan"].map(d)
df.head()
```

## Replace with `.replace() method

requires a dictionary with two elements: 
    
* Column to replace value in 
* Dictionary with replacement values

```python
df = df.replace({"Voice mail plan": d})
df.head()
```

# Grouping

`df.groupby(by = [grouping_columns])[columns_subset].function()`

```python
df.groupby(["Churn"])[["Total day minutes", "Total day calls"]].describe(percentiles = [])
```

By using `.agg([function1, function2, ...])` we can get different stats

```python
df.groupby(["Churn"])[["Total day minutes", "Total day calls"]]\
.agg([np.mean, np.std, np.min, np.max])
```

## Delete column with `.drop([cols])`

Requires to specify axis: *1* for columns and *0* for rows (default). 

Also there's an `inplace` argument - if `True`, the current dataframe is changed, otherwise the new one is created

```python
df.drop(["Total calls"], axis = 1, inplace=True)
```

# 2. Work on predicting telecom churn

```python
# explore data
pd.crosstab(df["Churn"], df["International plan"], margins = True, normalize=True)
```

```python
# Visualize
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = "svg"
```

```python
sns.countplot(x = "International plan", hue = "Churn", data = df)
```

```python
pd.crosstab(df.Churn, df["Customer service calls"], margins = True)
```

```python
sns.countplot(data = df, x = "Customer service calls", hue = "Churn")
```

```python
# let's add a binary feature `Customer service calls > 3` and compare it to churn
df["Many_service_calls"] = (df["Customer service calls"] > 3).astype("int")
pd.crosstab(df.Many_service_calls, df.Churn, margins = True)
```

```python
sns.countplot(data = df, 
             x = "Many_service_calls", 
             hue = "Churn")
```

```python
# we can build cross tab with combination of columns
pd.crosstab(df.Many_service_calls & df["International plan"], 
           df.Churn)
```

We can calculate accuracy of our prediction by summing values off main diagonal:
    
* **464** - we expected customer not to churn by they churned
* **9** - we expected them to churn but they did not

which gives us `(464 + 9)/3333 = 85.8%` accuracy
